```python
'''
importing all the required libraries
'''
import requests
from bs4 import BeautifulSoup
import csv
import numpy as np
import pandas as pd
```

```python
'''
get complete html content with thr help of beautifulSoup library
'''
source=requests.get("https://www.worldometers.info/coronavirus/").text
soup=BeautifulSoup(source,"html.parser")
```

```python
'''
scrape for the required tag and write it to the csv file
'''
table=soup.find_all("table",{"id":"main_table_countries_today"})
tbody=table[0].find("tbody")
with open('Worldo_meter.csv', 'w', newline='',encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["SLno","Country","Total cases","New cases","Total deaths","New deaths","Total recovered","Active cases","Serious critical","Total cases/1M pop","Deaths/1M pop","Total Tests","Tests/1M pop","Population","Continent"])
    for country in tbody.find_all("tr"):
        write=[]
        for one in country.find_all("td"):
            write.append(one.text)
        writer.writerow(write)
    
```

```python
'''
after writing into the csv file all the columns will be in object type.
but we need to convert it into interger for the visualization purpose.
'''
covid=pd.read_csv("Worldo_meter.csv")
covid.dropna(subset=["SLno"],how="any",inplace=True)
covid["Total cases"]=covid["Total cases"].str.replace(",","")
covid["Total cases"]=covid["Total cases"].astype("int64")

covid["Total deaths"]=covid["Total deaths"].str.replace(",","")
covid["Total deaths"]=covid["Total deaths"].str.replace(" ","")
covid.loc[covid["Total deaths"]=="","Total deaths"]=np.nan
covid["Total deaths"]=covid["Total deaths"].astype("Float32").astype("Int32")

covid["Active cases"]=covid["Active cases"].str.replace(",","")
covid["Active cases"]=covid["Active cases"].astype("Float32").astype("Int32")

covid["Population"]=covid["Population"].str.replace(",","")
covid["Population"]=covid["Population"].astype("int64")

covid["Total recovered"]=covid["Total recovered"].str.replace(",","")
covid["Total recovered"]=covid["Total recovered"].astype("Float32").astype("Int32")
```

```python
'''
convert csv to excel sheet. 
one can do visualization in csv also. this line of code is optional
'''
covid.to_excel("Worldo_meter.xlsx",index=False)
```
